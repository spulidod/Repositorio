Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np

mu = 0
beta = 0.5 
n=1000
vals = np.random.gumbel(mu, beta, size=n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad normal estándar.

```python
from scipy.stats import gumbel_r
import numpy as np
import matplotlib.pyplot as plt

# gumbel (miu=0, beta=0.5)
x= np.arange(-1,3,0.001)
plt.plot(x, gumbel_r.pdf(x,loc=0,scale=0.5))
plt.title('Función de densidad Gumbel')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad normal con media 3 y desviación estándar 1.

```python
from scipy.stats import gumbel_r
import numpy as np
import matplotlib.pyplot as plt


# gumbel (miu=1, beta=0.5)
x= np.arange(-0.5,4,0.001)
plt.plot(x, gumbel_r.pdf(x,loc=1,scale=0.5))
plt.title('Función de densidad Gumbel (mu=1, beta=0.5)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```
